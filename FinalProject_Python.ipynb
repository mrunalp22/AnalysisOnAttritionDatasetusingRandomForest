# Data Visualization using matplotlib and seaborn packages

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
def barPlot(feature):
    Yes = employee[employee['Attrition']==1][feature].value_counts()
    NO = employee[employee['Attrition']==0][feature].value_counts()
    df = pd.DataFrame([Yes,NO])
    df.index = ['Yes','NO']
    df.plot(kind='bar',stacked=True, figsize=(10,5))
    plt.show()
    print('Yes:\n',Yes)
    print('NO:\n', NO)
```

```python
barPlot('Gender')
```

The chart confirms that males are more likely to leave the jobs as compared to females.

```python
barPlot('MaritalStatus')
```

This chart confirms that employees with marital status single are more likely to leave the jobs as compared to employees with marital status Married whereas employees with marital status divorced are less likely to leave the jobs as compared to employees with marital status single and married.

```python
barPlot('OverTime')
```

This chart confirms that the employees who do not work overtime are not likely to leave the jobs as compared to employees who work overtime are less likely to not leave the jobs.

```python
barPlot('JobRole')
```

```python
def histo_plot(feature):
    histo = sns.FacetGrid(employee, hue='Attrition', size=5, aspect=3, palette='seismic')
    histo.map(plt.hist, feature ,bins=30,alpha=0.5,normed=True)
    histo.set(xlim=(0, employee.Age.max()+10))
    histo.add_legend()
```

```python
histo_plot('Age')
```

# Function to report results of Random Search

```python
# param_dist for randomized search
param_dist = {'max_depth':[3,5,10],
              'max_features':randint(1,11),
              'criterion':['gini','entropy'],
              'bootstrap':[True,False],
              'min_samples_leaf':randint(1,4),
              'min_samples_split':randint(2,11)
            }
```

```python
#Create function of cross validation
def hypertuning_rscv(rf, p_distr,n_iter, train_x, train_y):
    rdmsearch = RandomizedSearchCV(rf, param_distributions=p_distr,n_iter = n_iter, cv=8)
    rdmsearch.fit(train_x,train_y)
    ht_params = rdmsearch.best_params_
    ht_score = rdmsearch.best_score_
    return ht_params, ht_score
```

```python
rf_parameters, rf_ht_score = hypertuning_rscv(rf,param_dist,40,train_x,train_y )
```
